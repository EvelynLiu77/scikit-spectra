### ts.plot()

`TimeSpectra.plot()` is merely a wrapper to the pandas [dataframe.plot()](http://pandas.pydata.org/pandas-docs/dev/generated/pandas.DataFrame.plot.html) method.  It has some bells and whistles like colorbars, and default labels/legend settings. 

```python
from pyuvvis.data import aunps_glass

ts = aunps_glass(style=1)
ts.plot(cbar=True);
```

`ts.plot()` will reflect changes in the spectral, temporal and intensity units of the plot.  Let's now express the data as a time interval, and change the spectral unit.  We'll also look at the normalized data (relative to the first curve) by setting `ts.iunit = 'r'`

```python
t2 = ts.as_varunit('m')
t2 = t2.as_iunit('r')
t2.name = 'Normalized nanoparticle data.'
t2.plot(cbar=True, cmap='hot');
```

Likewise, we can compute the area directly from ts and plot it

```python
area = ts.area() #area is a hybrid array
plt.plot(area.transpose());
```

### normplot(), areaplot() and other convience plots

Because areaplots and referenced data (absorbance, transmittance etc...) are so important, pyuvvis includes a few simple wrapper plots to reduce the boilerplate.  For example, instead of bothering with `ts.area().transpose()`, we can just use `areaplot(ts)`.  This will give a more clear y-axis label, and title by default:

```python
from pyuvvis.plotting import areaplot

areaplot(ts);
```

#### Plotting Normalized Data

`pyuvvis` provides a shortcut keyword to plot data as Absorbance, Transmittance etc... see the [primary tutorial](http://nbviewer.ipython.org/github/hugadams/pyuvvis/blob/master/examples/Notebooks/tutorial_1.ipynb?create=1) for an explanation of these various representations.  

For now, let's plot the data in the **iunit='r'** representation, which simply is dividing all the curves in the set by the reference curve: in this case, that's the first spectrum (t=0).

```python
ts.plot(ts, iunit='r');
```

#### Ranged timeplot / Wavelength Slices

Another very common operation is slicing the data into subsets of wavelengths.  For example, one may ask how the curve varies between   ${\bf400nm \lt \lambda \le 450nm}$.  To simplify this, there is the TimeSpectra method `ts.wavelength_slices()` and the `range_timeplot` (or stripchart as its sometimes called).

```python
from pyuvvis.plotting import range_timeplot

ts_by_8 = ts.wavelength_slices(8)  # Eight even slices
range_timeplot(ts_by_8);
```

### Multi Plots

Pyuvvis has builtin multiplot support, but is designed to be flexible in creating arbitrary multiplots.  One such builtin plot is the `quadplot()`, which essentially summarized the plots seen so far:

```python
from pyuvvis.plotting import quad_plot

fig = quad_plot(ts, striplegend=True) #Show strip chart wavelength legend:
                                               #help(quad_plot) for other options
```

Let's split the data into 5 sets and show this as a multiplot.  We can easily split the data using `ts.splitby(n, axis=1)`. We'll also normalize the data using `as_iunit('r')`  Sampling will automatically store subsets in a pyuvvis.SpecStack object; see the [SpecStack tutorial](http://nbviewer.ipython.org/github/hugadams/pyuvvis/blob/master/examples/Notebooks/specstack.ipynb?create=1) for more help.

```python
stack = ts.split_by(5)
stack.iunit = 'r'
stack.reference = 0  # FLIP THESE FOR BUG

stack.plot(figsize=(8,8), colormap='RdBu');
```

Notice that 5 subplots were created for us automatically.  You can specify the number of columns to `SpecStack.plot()` through the `ncols` keyword (default =3).  e.g. `stack.plot(ncols=2)`

### Custom multiplots

*Custom multiplots are easy because pyuvvis and pandas plots take references to any maplotlib Axes object.*  You can generate empty Axes via `plt.subplots()` or through the pyuvvis subplots wrapper `splot`.  splot is almost identicaly to plt.subplots(), but does not require you to nest your axes calls.  For example, a (2,3) subplot could be generated in the following ways:

- `fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2,3)`
- `ax1, ax2, ax3, ax4, ax5, ax6 = splot(2,3)`

I often find this nesting superfulous for my tasks.  `splot` has some other features (*help(splot)*) not described here.

**Let's use splot to compare wavelengths regions of the spectra**

```python
from pyuvvis.plotting import splot

ax1, ax2 = splot(1,2)

# Wavelength slices (up to 500, 600 on)
ts[:500.0].plot(ax=ax1, title='Short Wavelengths', color='b')
ts[600.0:].plot(ax=ax2, title='Long Wavelengths', color='r')

ax2.yaxis.set_label_position('right')
```

As one last example, let's show off this data set, normalized by the first curve, in four representations: transmittance, absorbance (base 10), absorbance (base e), and percent transmittance.

```python
ax1, ax2, ax3, ax4 = splot(2,2, figsize=(8,8))

ts.plot(ax=ax1, iunit='t')
ts.plot(ax=ax2, iunit='a')
ts.plot(ax=ax3, iunit='a(ln)', title='')
ts.plot(ax=ax4, iunit='%t', title='');

ax2.yaxis.set_label_position('right')
ax4.yaxis.set_label_position('right')
```
