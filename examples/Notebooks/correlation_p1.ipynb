### Synchronous and Asyncronous Spectra

Let's start with some simple test data of protein Bovine Serum Albumin binding to gold nanopaticles in water.  We'll work in **Absorbance** units.

```python
from pyuvvis.data import aunps_water
ts = aunps_water().as_iunit('a').ix[220.0:700.0, :] #Slice wavelengths 220-700
ts.plot(cbar=True, colormap='bone');
```

The spectra change at wavelengths 200-300nm is the uv absorbance of the protein, while the change at 500-700nm is due to the plasmon resonance of the gold particles.  Let's initiate our correlation analysis using this dataset through the **pyuvvis.core.corr.Corr2D class**.  **<font color='blue'>This is designed to work with numpy arrrays for maximum reuse</font>,** but since we already have a Spectra, let's go ahead and instantiate it using the **from_method()** constructor.

```python
from pyuvvis import Corr2d
cspec = Corr2d.from_spectra(ts)
cspec.plot(cbar=True)
cspec;
```

We can look at the *asynchronous* spectra to see which regions of the spectra are changing independently; ie, uncorrelated.

```python
cspec.plot('asynchronous', cbar=True, cmap='cool', fill=True);
```
