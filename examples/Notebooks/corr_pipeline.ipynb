### Load the builitin dataset from [1] that depicts solvent evaporation

```python
from pyuvvis.data import solvent_evap

help(solvent_evap)
```

#### Plot the data, and denote time through a [Two-channel colormap](http://wiki.scipy.org/Cookbook/Matplotlib/Show_colormaps)

```python
ts = solvent_evap()
ts.plot(cbar=True, colormap='spring', padding=0);
```

### Time and Spectral Slicing

With the above colormap, we can begin to see the order of events in the experiment.  Let's choose three regions of interest to focus on.  A good approach is to use plotly to interactively zoom in on separate regions in the dataset.   

```python
import plotly.plotly as py

fig, (ax1, ax2, ax3) = plt.subplots(1,3)
ts.plot(ax=ax1, colormap='spring', title='2700-3200')
ts.plot(ax=ax2, colormap='spring')
ts.plot(ax=ax3, colormap='spring')
py.iplot_mpl(fig)
```

```python
from pyuvvis.plotting import areaplot
areaplot(ts)
```
